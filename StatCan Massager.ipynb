```python
# coding: utf-8

# Processess figures from the massive Statistics Canada Census Profile csv
# into a form suitable to be joined to GIS data for the relevant
# geographic levels (code below uses Dissemination Areas)
# 
# Nicole White February 2019

import pandas as pd
import numpy as np
import os

# This will hold the final output df
result = pd.DataFrame()

# Get the raw data from the csv as a df.
df = pd.read_csv('D:\\Winter2019\\GIS2\\Project\\data\\StatCan\\ns.csv', low_memory=False)
df = df.loc[df['GEO_LEVEL'] == 4]

# Get rid of these columns
fields_to_drop = ['GNR', 'GNR_LF', 'DATA_QUALITY_FLAG','ALT_GEO_CODE', 'Member ID: Profile of Dissemination Areas (2247)', 'Notes: Profile of Dissemination Areas (2247)', 'Dim: Sex (3): Member ID: [2]: Male', 'Dim: Sex (3): Member ID: [3]: Female', 'GEO_NAME', 'GEO_LEVEL', 'CENSUS_YEAR']
df = df.drop(fields_to_drop, axis=1)

# Get list of DAs of interest
with open('D:\\Winter2019\\GIS2\\Project\\data\\StatCan\\das_to_get2.csv') as i:
    my_das = [int(x.strip('\n').strip('"').strip('ï»¿')) for x in i.readlines()]

new_das = []
new_df = pd.DataFrame()

# Unneeded fields that will be dropped, stored in csv form
# (called 'langfields' because most are language names, but not all)
with open('D:\\Winter2019\\GIS2\\Project\\data\\StatCan\\langfields.csv') as i:
    langfields = [x.strip('\n').strip('"').strip('ï»¿') for x in i.readlines()]

print('Checking')

# Loop through the list of DAs to be retrieved. If the df does indeed
# contain them, save them to a new list.
for x, da in enumerate(my_das):
    da = str(da)
    if df['GEO_CODE (POR)'].astype(str).str.contains(da).any():
        new_das.append(da)
    else:
        print('Not found: {}'.format(da))
print('Done!')
```

```python
print('Processing')
for da in new_das:
    new_df = df.loc[df['GEO_CODE (POR)'] == int(da)]
    new_df = new_df.transpose()
    headers = new_df.iloc[1]
    new_df = new_df.rename(columns = headers)
    new_df = new_df.drop(['DIM: Profile of Dissemination Areas (2247)','GEO_CODE (POR)'], axis=0)
    new_df['dauid'] = da
    new_df = new_df.drop(langfields, axis=1)
    
    # if the dataframe called result is empty, result = new_df
    if len(result) == 0:
        new_df.to_csv('D:\\Winter2019\\GIS2\\Project\\data\\StatCan\\new_df1.csv')
        result = new_df
    # if result is not empty, append new_df to it
    else:
        new_df.to_csv('D:\\Winter2019\\GIS2\\Project\\data\\StatCan\\new_df2.csv')
        result = result.append(new_df)
        
# Save to csv. Any further desired editing can be done in the GIS software
result.to_csv('D:\\Winter2019\\GIS2\\Project\\data\\StatCan\\result.csv')
print('Done!')
```
